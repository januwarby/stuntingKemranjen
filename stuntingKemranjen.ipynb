```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://github.com/januwarby/stuntingKemranjen/raw/master/BADUTA-BATITA%20KEMRANJEN.xlsx'
df = pd.read_excel(url)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Menghitung persentase nilai null dalam dataframe
null_percentage = df.isnull().mean() * 100

# Membuat dataframe baru dari persentase nilai null
null_df = pd.DataFrame({'Column': null_percentage.index, 'Null Percentage': null_percentage.values})

null_df
```

```python
# Mengisi hanya kolom numerik dengan nilai mean
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
df.isnull().sum()
```

```python
df_cleaned = df.dropna()
```

```python
df_cleaned.isnull().sum()
```

```python
columns_to_drop = ['NamaBaduta', 'KodeSampel', 'Alamat', 'RT', 'RW', 'NamaAyah', 'NamaIbu','TanggalLahir','TanggalUkur']
df_cleaned = df.drop(columns_to_drop, axis=1)
```

```python
df_cleaned.info()
```

```python
df_cleaned['JenisKelamin'] = df_cleaned['JenisKelamin'].replace({'Laki-laki': 1, 'Perempuan': 0})
df_cleaned['Status'] = df_cleaned['Status'].replace({'stunting': 1, 'tidak stunting': 0})
```

```python
df_cleaned.head()
```

```python
# Menghitung korelasi antar kolom
correlation = df_cleaned.corr()

# Membuat heatmap korelasi
plt.figure(figsize=(50, 40))
sns.heatmap(correlation, annot=True, cmap='coolwarm', square=True, vmin=-1, vmax=1)

# Menampilkan heatmap
plt.title('Heatmap Korelasi')
plt.show()
```

```python
columns_to_drop1 = ['Status_ZscoreBBPB', 'Status_ZscoreIMTU', 'Status_ZscoreBBU', 'Status_ZscorePBU']
df_clean = df_cleaned.drop(columns_to_drop1, axis=1)
```

```python
df_clean.info()
```

```python
# Menghitung persentase kategori dalam kolom "DEATH_EVENT"
death_event_counts = df_clean['JenisKelamin'].value_counts()
percentages = death_event_counts * 100 / death_event_counts.sum()

# Daftar warna untuk setiap kategori
colors = ['skyblue', 'salmon']

# Membuat diagram batang dengan warna yang berbeda
fig, ax = plt.subplots()
bars = ax.bar(death_event_counts.index.astype(str), death_event_counts, color=colors)

# Menambahkan keterangan persentase di atas setiap batang
for bar, percentage in zip(bars, percentages):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height,
            f'{percentage:.1f}%', ha='center', va='bottom')

# Memberi judul diagram dan label sumbu
ax.set_title('Status Distribution')
ax.set_xlabel('Status')
ax.set_ylabel('Count')

# Menampilkan diagram
plt.show()
```
